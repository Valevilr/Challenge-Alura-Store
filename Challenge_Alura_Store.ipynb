### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Facturación total por tienda

fact_t1 = tienda["Precio"].sum()
fact_t2 = tienda2["Precio"].sum()
fact_t3 = tienda3["Precio"].sum()
fact_t4 = tienda4["Precio"].sum()

print("Facturación Tienda 1:", fact_t1)
print("Facturación Tienda 2:", fact_t2)
print("Facturación Tienda 3:", fact_t3)
print("Facturación Tienda 4:", fact_t4)
```

```python
import matplotlib.pyplot as plt

facturaciones = [fact_t1, fact_t2, fact_t3, fact_t4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, facturaciones)
plt.title("Facturación Total por Tienda")
plt.ylabel("Total en Ventas")
plt.show()
```

# 2. Ventas por categoría

```python
print("Tienda 1")
print(tienda["Categoría del Producto"].value_counts())

print("\nTienda 2")
print(tienda2["Categoría del Producto"].value_counts())

print("\nTienda 3")
print(tienda3["Categoría del Producto"].value_counts())

print("\nTienda 4")
print(tienda4["Categoría del Producto"].value_counts())
```

```python
tienda["Categoría del Producto"].value_counts().plot(kind="bar")
plt.title("Categorías - Tienda 1")
plt.show()
```

# 3. Calificación promedio de la tienda


```python
cal_t1 = tienda["Calificación"].mean()
cal_t2 = tienda2["Calificación"].mean()
cal_t3 = tienda3["Calificación"].mean()
cal_t4 = tienda4["Calificación"].mean()

print("Promedio Tienda 1:", cal_t1)
print("Promedio Tienda 2:", cal_t2)
print("Promedio Tienda 3:", cal_t3)
print("Promedio Tienda 4:", cal_t4)
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure()
plt.plot(tiendas, [cal_t1, cal_t2, cal_t3, cal_t4])
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación Promedio")
plt.show()
```

# 4. Productos más y menos vendidos

```python
tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

for nombre, df in tiendas.items():
    print(f"\n {nombre} - Producto más vendido")
    print(df["Producto"].value_counts().head(1))

    print(f"\n {nombre} - Producto menos vendido")
    print(df["Producto"].value_counts().tail(1))
    print("\n" + "="*40)

```

# 5. Envío promedio por tienda

```python
env_t1 = tienda["Costo de envío"].mean()
env_t2 = tienda2["Costo de envío"].mean()
env_t3 = tienda3["Costo de envío"].mean()
env_t4 = tienda4["Costo de envío"].mean()

print("Costo promedio envío Tienda 1:", env_t1)
print("Costo promedio envío Tienda 2:", env_t2)
print("Costo promedio envío Tienda 3:", env_t3)
print("Costo promedio envío Tienda 4:", env_t4)
```

```python
plt.figure()
plt.scatter(facturaciones, [env_t1, env_t2, env_t3, env_t4])

for i, txt in enumerate(tiendas):
    plt.annotate(txt, (facturaciones[i], [env_t1, env_t2, env_t3, env_t4][i]))

plt.title("Relación entre Facturación y Costo de Envío")
plt.xlabel("Facturación")
plt.ylabel("Costo Promedio de Envío")
plt.show()
```

# 6. Desempeño geográfico (extra)

```python
import numpy as np

plt.figure(figsize=(10,8))

# Pequeño desplazamiento visual
offset = 0.05

plt.scatter(tienda["lon"] - offset, tienda["lat"] - offset,
            s=15, alpha=0.6, c="blue", label="Tienda 1")

plt.scatter(tienda2["lon"] - offset, tienda2["lat"] + offset,
            s=15, alpha=0.6, c="red", label="Tienda 2")

plt.scatter(tienda3["lon"] + offset, tienda3["lat"] - offset,
            s=15, alpha=0.6, c="green", label="Tienda 3")

plt.scatter(tienda4["lon"] + offset, tienda4["lat"] + offset,
            s=15, alpha=0.6, c="orange", label="Tienda 4")

plt.title("Desempeño Geográfico Comparativo - Alura Store LATAM", fontsize=14)
plt.xlabel("Longitud")
plt.ylabel("Latitud")

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()





```

Las tiendas presentan una distribución geográfica similar, lo que indica que operan en las mismas regiones. No se observan diferencias significativas en cobertura territorial, por lo que el desempeño no parece estar determinado por ubicación geográfica sino por otros factores como ingresos, calificaciones o costos.

# **INFORME FINAL**

El presente análisis tiene como objetivo evaluar el desempeño de las cuatro tiendas de Alura Store LATAM con el fin de determinar cuál de ellas debería ser vendida por el Sr. Juan. Para ello, se analizaron diversos indicadores clave, tales como los ingresos totales, las categorías de productos más y menos vendidas, las calificaciones promedio de los clientes, los productos con mayor y menor rotación, el costo promedio de envío y el desempeño geográfico.

A través de este estudio se busca identificar fortalezas y debilidades en cada tienda para tomar una decisión estratégica basada en datos.                  

##Desarrollo del análisis:

###Ingresos totales

Al analizar la facturación de cada tienda, se observa que la Tienda 4 presenta el menor nivel de ingresos en comparación con las demás. Este indicador clave, ya que la facturación refleja directamente el rendimiento comercial y la capacidad de generación de ingresos.

Las tiendas 1, 2 y 3 muestras un desempeño superior en términos de ventas totales.

###Ventas por categoria

El análisis por categoria permitió cuáles son los tipos de productos con mayor demanda en cada tienda. En general, algunas categorías concentran la mayor parte de las ventas, lo que indica una preferencia clara de los clientes.

Sin embargo, la Tienda 4 muestra menro volumen en varias categorías relevantes, lo que impacta directamente en su facturación total.

###Calificación promedio de los clientes

En términos de satisfacción del cliente, se observó que la Tienda 1 presenta la calificación promedio más baja (3.97), lo que podría represnetar un área de mejora.

No obstante, las diferencias entre tiendas no son extremadamente amplias, por lo que este factor, aunque importate, no resulta determinante frente a la diferencia en ingresos.

###Productos más y menos vendidos

El análisis de productos permitió identificar aquellos con mayor rotación y aquellos con bajo desempeño. Las tiendas con mayores ingresos presentan productos con alta demanda constante, mientras que la Tienda 4 muestra menor volumen incluso en sus productos más vendidos.

Esto refuerza la evidencia de un desempeño comercial inferior.

###Costo de envío promedio

El análisis del costo de envío indica que la Tienda 1 presenta el envío promedio más alto (26018.60). Aunque este costo es asumido por el cliente, puede influir en la percepción de valor y competitividad.

Sin embargo, este factor no compensa la diferencia significativa en ingresos observada en la Tienda 4.

###Desempeño geográfico

El análisis geográfico mostró que las cuatro tiendas operan en regiones similares, sin diferencias significativas en cobertura territorial. Esto sugiere que la ubicación no es el factor principal que explica el bajo desempeño de la Tienda 4.

Por lo tanto, las diferencias observadas parecen estar relacionadas con factores comerciales internos y no con limitaciones geográficas.

##Conclusión y recomendación final:

Con base en los análisis realizados, se recomienda que el Sr. Juan considere la venta de la Tienda 4.

Esta recomendación se fundamenta principalmente en:

La menor facturación total entre las cuatro tiendas.

Menor desempeño en diversas categorías de productos.

Menor volumen de ventas en comparación con las demás tiendas.

Ausencia de factores geográficos que justifiquen su bajo rendimiento.

Aunque la Tienda 1 presenta la calificación promedio más baja y el mayor costo de envío, su nivel de ingresos es superior al de la Tienda 4, lo que la convierte en una unidad comercial más sólida.

En conclusión, la Tienda 4 presenta el desempeño general más bajo y, desde una perspectiva estratégica y financiera, es la opción más razonable para ser vendida.
